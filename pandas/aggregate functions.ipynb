```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df = sns.load_dataset("titanic")
```

```python
df.head(3)
```

```python
# fonkisyon uygulama 
# aggregate fonksiyonu
df.dtypes
```

```python
df_number = df.select_dtypes(include=["float64","int64"])  # belli veri tipini seçtik. 
df_number.head()
```

```python
df_number.aggregate(["min","max"])  # bütün sütünların içinde bulunan max ve min değerler
```

```python
df["age"].min()
```

```python
df_number.aggregate([np.median])  # medyan değerlerini ekran yazdırdık
```

```python
df_number.aggregate(["mean","median"])
```

```python
df_number.aggregate(["mean",np.median])
```

```python
df_number.aggregate({"survived":"mean","pclass":min,"age":max})  # sözlük yapısı olarak da verileri çekebiliriz
```

```python
def f(x):
    return x.mean()-x.std()
```

```python
df_number.aggregate(f)  # 27 ve 28. hücreler aynı
```

```python
f(df_number)   # 27 ve 28. hücreler aynı
```

```python
df_number.aggregate(["std","mean",f])
```

```python
fonksiyon = ["max","mean","std"]
```

```python
df_number.aggregate(fonksiyon)
```
